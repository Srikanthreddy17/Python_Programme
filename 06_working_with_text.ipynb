## Splitting the strings

```python
customers = pd.read_csv("customers.csv") 
```

```python
customers
```

```python
customers["Name"].str.len()
```

```python
customers["Name"].str.split("")
customers["Name"].str.split(pat=" ")
```

```python
customers["Name"].str.split(pat=" ", n=1)
```

```python
customers["Name"].str.split(pat=" ", n=1, expand= True)
```

```python
customers
```

```python
customers[["First Name","Last Name"]]= customers["Name"].str.split(pat= " ", n=1, expand= True)
```

```python
customers
```

```python
customers = customers.drop(labels="Name", axis="columns")
```

```python
customers
```

```python
customers["Address"].str.split(pat=",", n=4, expand= True)
```

```python
customers[["Street","City","State","Pincode"]] = customers["Address"].str.split(pat=",", n=4, expand= True)
```

```python
customers
```

```python
customers = customers.drop(labels="Address", axis="columns")
```
