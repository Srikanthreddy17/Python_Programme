```python
employees = pd.read_csv("employees.csv", parse_dates=["Start Date"])
```

```python
employees
```

```python
employees["Mgmt"] = employees["Mgmt"].astype(bool)
```

```python
# employees["Salary"].astype(int)
```

```python
employees["Salary"] = employees["Salary"].fillna(0).astype(int)
```

```python
employees.dtypes
```

```python
employees["Team"] = employees["Team"].astype("category")
```

```python
employees.dtypes
```

## Filtering the DataFrame by Colomns

```python
higher_than_80k = employees["Salary"]>8000
higher_than_80k
```

```python
employees[higher_than_80k]
```

```python
employees[employees["Mgmt"]] # ~ is the inversion operator
```

```python
employees[~employees["Mgmt"]]
```

```python
after_2015 = employees["Start Date"]>"2015-01-01"
employees[after_2015]
```

```python
jan_2015 = employees["Start Date"].between(
    left = "2015-01-01",
    right = "2015-07-31"
)
employees[jan_2015]
```

```python
all_star_teams = ["Sales","Marketing","Finance"]
star_teams = employees["Team"].isin(all_star_teams)
employees[star_teams]
```

```python
employees["Team"].value_counts()
```

```python
starting_with_s = employees["First Name"].between("S","T")
employees[starting_with_s] 
```
